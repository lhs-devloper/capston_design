```python
stats_list = [
        "Unnamed: 0",
        "year",
        "team",
        "age",
        "G",
        "PA",
        "AB",
        "R",
        "HIT",
        "HIT_2",
        "HIT_3",
        "HOME_RUN",
        "BASE",
        "RBI",
        "SB",
        "CS",
        "HBP",
        "IBB",
        "SO",
        "GDP",
        "SH",
        "SF",
        "AVG",
        "OBP",
        "SLG",
        "OPS",
        "wOBA",
        "wRC",
        "WAR",
        "Name",
    ]
df1 = pd.DataFrame(columns=stats_list)
df1.head()
```

```python
for i in player_list:
    df = pd.read_csv(f"{i}")
    if len(df)<10:
        continue
    name = i.replace('re_data\\','')
    name = name.replace('.csv','')
    df["Name"] = name
    df1 = pd.concat([df1, df])
    
    
```

```python
df1 = df1.drop(["Unnamed: 0"],axis=1)
df1 = df1.drop(['Unnamed: 0.1'], axis=1)
df1 = df1[df1["year"]<8]
df1[df1["Name"]=="이대호"]
```

```python
# df1.to_csv("save.csv")
df1.columns
```

```python
def fill_float(WAR):
    if WAR==" ":
        return 0
    elif WAR==0:
        return 0
    else:
        return float(WAR)
def fill_int(WAR):
    if WAR==" ":
        return 0
    elif WAR==0:
        return 0
    else:
        return int(WAR)
df1["WAR"] = df1["WAR"].apply(fill_float)
df1["year"] = df1["year"].apply(fill_int)
df1["age"] = df1["age"].apply(fill_int)
df1["G"] = df1["G"].apply(fill_int)
df1["PA"] = df1["PA"].apply(fill_int)
df1["AB"] = df1["AB"].apply(fill_int)
df1["R"] = df1["R"].apply(fill_int)
df1["HIT"] = df1["HIT"].apply(fill_int)
df1["HIT_2"] = df1["HIT_2"].apply(fill_int)
df1["HIT_3"] = df1["HIT_3"].apply(fill_int)
df1["HOME_RUN"] = df1["HOME_RUN"].apply(fill_int)
df1["BASE"] = df1["BASE"].apply(fill_int)
df1["RBI"] = df1["RBI"].apply(fill_int)
df1["SB"] = df1["SB"].apply(fill_int)
df1["CS"] = df1["CS"].apply(fill_int)
df1["HBP"] = df1["HBP"].apply(fill_int)
df1["IBB"] = df1["IBB"].apply(fill_int)
df1["SO"] = df1["SO"].apply(fill_int)
df1["SH"] = df1["SH"].apply(fill_int)
df1["SF"] = df1["SF"].apply(fill_int)
#---------------------------------------------------
df1["AVG"] = df1["AVG"].apply(fill_float)
df1["OBP"] = df1["OBP"].apply(fill_float)
df1["SLG"] = df1["SLG"].apply(fill_float)
df1["OPS"] = df1["OPS"].apply(fill_float)
df1["wOBA"] = df1["wOBA"].apply(fill_float)
df1["wRC"] = df1["wRC"].apply(fill_float)
```

```python
def type_first(stats):
    if stats>0.47:
        return "Slugger"
    else:
        return "Balance"
def type_second(stats):
    if stats>14:
        return "Speed"
    else:
        return "Balance"
df1["type"] = df1["SLG"].apply(type_first)
df1["type"] = df1["SB"].apply(type_second)
```
